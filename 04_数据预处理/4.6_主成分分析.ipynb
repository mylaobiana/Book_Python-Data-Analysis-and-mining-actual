```python
import pandas as pd
inputFile = '.\chapter4\demo\data\principal_component.xls'
outputFile = './dimention_reducted.xls'
data = pd.read_excel(inputFile, header=None)
data
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(data)
```

```python
pca.components_  # 返回模型的各个特征向量
```

```python
pca.explained_variance_ratio_  # 返回各个成分各自的方差百分比（贡献率）
```

```python
# 设置 3 个主要成分
pca = PCA(n_components=3)
pca.fit(data)
low_d = pca.transform(data)  # 降低维度
pd.DataFrame(low_d).to_excel(outputFile)  # 保存结果
pca.inverse_transform(low_d)  # 复原数据
```
